```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Lista de localidades.xlsx', header = 2)
data
```

```python
import re

#Convert the data to a DataFrame
df = pd.DataFrame(data)

# Function to split the location and the number, handling thousands separators
def split_location_number(cell):
    match = re.match(r'^(.*?)(\d{1,3}(?:\.\d{3})*)$', str(cell).strip())
    if match:
        location = match.group(1).strip()
        number = match.group(2).replace('.', '').strip()
        return location, number
    else:
        return cell, None

# Process each column
new_columns = {}
for column in df.columns:
    locations = []
    numbers = []
    for cell in df[column]:
        location, number = split_location_number(cell)
        locations.append(location)
        numbers.append(number)
    
    new_columns[f'{column}_Location'] = locations
    new_columns[f'{column}_Number'] = numbers
```

```python
# Create a new DataFrame with the processed columns
new_df = pd.DataFrame(new_columns)

new_df
```

```python
# Save the new DataFrame to an Excel file
new_df.to_excel('output.xlsx', index=False)
```
